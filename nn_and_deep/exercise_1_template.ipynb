## 1 Basic definitions

```python
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt
```

#### (a)

```python
class Net(nn.Module):
    def __init__(self, activation='relu'):
        super().__init__()
        self.layers = nn.Sequential(
            nn.Linear(5, 5),
            self.get_activation(activation),
            nn.Linear(5, 5),
            self.get_activation(activation),
            nn.Linear(5, 5)
        )

    def get_activation(self, activation):
        if activation == 'relu':
            return nn.ReLU()
        elif activation == 'tanh':
            return nn.Tanh()
        elif activation == 'linear':
            return nn.Identity()
        else:
            raise ValueError("Invalid ")


    def forward(self, x):
        x = torch.flatten(x, 1)
        x = self.layers(x)
        return x

model = Net(activation='relu')
print(model)
```

#### (b)

```python
torch.manual_seed(0)
activations = ['relu', 'tanh', 'linear']
weights = [Net(activation=act) for act in activations]

# Code for reporting the 9 plots:
fig, axs = plt.subplots(nrows=3, ncols=3, figsize=(15, 10))
for i in range(3): # loop over rows of plot
    activation = activations[i]
    model = weights[i]
    model.apply(lambda m: m.reset_parameters() if hasattr(m, 'reset_parameters') else None)
    x1_values = np.linspace(-10.0, 10.0, 100)
    for j in range(3): # loop over cols of plot
        random_x_rest = np.random.randn(4)
        X = np.zeros((100, 5))
        X[:, 0] = x1_values
        X[:, 1:] = random_x_rest
        X_torch = torch.from_numpy(X).float()
        with torch.no_grad():
            outputs = model(X_torch)
        y1_values = outputs[:, 0].numpy()
        axs[i, j].plot(x1_values, y1_values)
        axs[i, j].set_title(f'activation {activation} {j+1}')
        axs[i, j].set_xlabel('x1')
        axs[i, j].set_ylabel('y1')
plt.tight_layout()
plt.show()
```
